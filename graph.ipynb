```python
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline  
```

```python
f = open('./result')
f2 = open('./feature')

name = []
for line in f:
    name.append(line.split('     ')[0])
#print name

f.close()

feature = []
for line in f2:
    temp = map(int, line.replace('[','').replace(']','').split(',')[:])
    feature.append(temp)
#print feature    
f2.close()
```

```python
G=nx.Graph()

x = 0

for obj in feature:
    y = 0
    for element in obj:
        if element != 0 and element != -1:
            G.add_edge(x, y, weight = element)
        y = y + 1
    x = x + 1
```

```python
nx.draw(G)
plt.show()
```
