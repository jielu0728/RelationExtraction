```python
from gensim.models import word2vec
import numpy as np
```

```python
sentences = word2vec.Text8Corpus('./textdone')
model = word2vec.Word2Vec(sentences, size=100, window=20, min_count=1)
```

```python
f = open('./result_final')
name = []
for line in f:
    name.append(line.split('      ')[0])
    
print name
```

```python
feature = []
for obj in name:
    try:
        feature.append(model[obj])
    except KeyError:
        for c in name:
            if '_' in c and '_' in obj and obj in c and c != obj:
                print obj+' and '+c+' are probably the same person.'
                feature.append(model[c])
print feature
```
